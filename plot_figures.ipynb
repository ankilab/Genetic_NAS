```python
import matplotlib.pyplot as plt
import seaborn as sns

with open('results_14032023164450/generation_0/fitness_list.pkl', 'rb') as f:
    fitness_0 = pickle.load(f)
    f.close()
fitness_0 = [i for i in fitness_0 if i != 0]
with open('results_14032023164450/generation_6/fitness_list.pkl', 'rb') as f:
    fitness_6 = pickle.load(f)
    f.close()
fitness_6 = [i for i in fitness_6 if i != 0]

sns.set_style('whitegrid')

fig, ax = plt.subplots()
boxprops = dict(linestyle='-', color='darkgray')
medianprops = dict(linestyle='-', color='navy')
whiskerprops = dict(linestyle='--', color='gray')

bp = ax.boxplot([fitness_0, fitness_6], boxprops=boxprops, medianprops=medianprops, whiskerprops=whiskerprops, showfliers=False, patch_artist=True)
colors = ['lightblue', 'cornflowerblue']
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)

ax.set_xticklabels(['Generation 0', 'Generation 6'])
ax.set_ylabel('Fitness')
ax.tick_params(axis='both')
ax.set_title('Comparison of Fitness')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


with open('results_14032023164450/generation_0/tpu_time_list.pkl', 'rb') as f:
    speed_0 = pickle.load(f)
    f.close()
speed_0 = [i for i in speed_0 if i != 9999]
with open('results_14032023164450/generation_6/tpu_time_list.pkl', 'rb') as f:
    speed_6 = pickle.load(f)
    f.close()
speed_6 = [i for i in speed_6 if i != 0]

sns.set_style('whitegrid')

fig, ax = plt.subplots()
boxprops = dict(linestyle='-', color='darkgray')
medianprops = dict(linestyle='-', color='navy')
whiskerprops = dict(linestyle='--', color='gray')

bp = ax.boxplot([speed_0, speed_6], boxprops=boxprops, medianprops=medianprops, whiskerprops=whiskerprops, showfliers=False, patch_artist=True)
colors = ['lightblue', 'cornflowerblue']
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)

ax.set_xticklabels(['Generation 0', 'Generation 6'])
ax.set_ylabel('Inference Speed (ms)')
ax.tick_params(axis='both')
ax.set_title('Comparison of Inference Speed')
plt.show()
```

```python

import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

plt.rcParams.update({"figure.dpi": 200})
```

```python
with open('results_14032023164450/max_fitness_history.pkl', 'rb') as f:
    a1 = pickle.load(f)
    f.close()
with open('results_28032023195907/max_fitness_history.pkl', 'rb') as f:
    a2 = pickle.load(f)
    f.close()
with open('results_29032023085605/max_fitness_history.pkl', 'rb') as f:
    a3 = pickle.load(f)
    f.close()
max_fitness_history = a1 + a2 + a3


with open('results_14032023164450/average_fitness_history.pkl', 'rb') as f:
    b1 = pickle.load(f)
    f.close()
with open('results_28032023195907/average_fitness_history.pkl', 'rb') as f:
    b2 = pickle.load(f)
    f.close()
with open('results_29032023085605/average_fitness_history.pkl', 'rb') as f:
    b3 = pickle.load(f)
    f.close()

average_fitness_history = b1 + b2 + b3

generations = range(19)

fig, ax = plt.subplots()

ax.step(generations, max_fitness_history, label='Max Fitness', where='post', color='red', linewidth=2)
ax.step(generations, average_fitness_history, label='Average Fitness', where='post', color='blue', linewidth=2)
ax.xaxis.set_major_locator(ticker.MultipleLocator(4))
ax.set_xlim(-1,19)
ax.set_xlabel('Generations', fontsize=12)
ax.set_ylabel('Fitness', fontsize=12)
ax.legend(fontsize=12)

plt.tight_layout()
plt.show()
```
